***Lấy mẫu theo cụm***
* Lấy mẫu với kích thước 30, điều kiện:
 *  Chia tổng thể thành 5 phần. mỗi phần chọn 6 cá thể làm mẫu.


```python
import numpy as np
# Chọn 5 SBD để chia thành 5 khoảng và lấy 6 cá thể trong mỗi khoảng
j=33000001
clusters=[]
for i in (33002500,33005000,33007500,33010000,33013345):
  x = np.random.choice(np.arange(j,i), size=6,replace=False)
  for k in x:
    clusters.append(k) # Danh sách chứa SBD đã được chọn.
  j=i
print(clusters)
```

```python
cluster_sample = data[data['SBD'].isin(clusters)]
cluster_sample
```

```python
# Kiểm tra kích thước mẫu
cluster_sample['SBD'].shape
```

```python
freq_toan=data['Toan'].value_counts()
freq_toan
```

***Biểu đồ thể hiện***

```python
import matplotlib.pyplot as plt
plt.bar(freq_toan.index,freq_toan.values)
plt.title('Phổ điểm thi môn toán THPT 2021')
plt.xlabel('Math')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(freq_toan.index,freq_toan.values)
plt.title('Phổ điểm thi môn toán THPT 2021')
plt.xlabel('Math')
plt.ylabel('Frequency')
plt.show()
```
